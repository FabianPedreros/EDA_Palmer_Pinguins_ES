## <a id='4.4'>4.4 Revisión de valores extremos</a>

```python
# Revisión de valores extremos para establecer outliers y si esos registros deben ser utilizados o no en el EDA.
# Se utilizan dos metodologías para establcer outliers:
# el Z Score como medida que tiene en cuanta la media
# y el Modified Z Score como medida que uriliza la mediana

# Z-Score
# Culmen Length (mm)
df2 = df.rename(columns={"Culmen Length (mm)":"CL","Culmen Depth (mm)":"CD","Flipper Length (mm)":"FL","Body Mass (g)":"BM","Delta 15 N (o/oo)":"N","Delta 13 C (o/oo)":"C"})
df2.columns
_, mean_CL, std_CL, *_ = df2.CL.describe()
def get_z_score_CL(value, mean_CL, std_CL):
    return(value - mean_CL)/std_CL
df2["z_score_CL"] =df2.CL.apply(lambda x: get_z_score_CL(x,mean_CL,std_CL))
df2.head()
```

```python
df2[df2.z_score_CL>3]
```

```python
# No se encuentran valores cuyo Z-Score sea mayor a tres veces, por lo que no existen outliers para el Culmen Depth
```

```python
# Culmen Depth (mm)
_, mean_CD, std_CD, *_ = df2.CD.describe()
def get_z_score_CD(value, mean_CD, std_CD):
    return(value - mean_CD)/std_CD
df2["z_score_CD"] =df2.CD.apply(lambda x: get_z_score_CD(x,mean_CD,std_CD))
df2.head()
```

```python
df2[df2.z_score_CD>3]
```

```python
# No se encuentran valores cuyo Z-Score sea mayor a tres veces, por lo que no existen outliers para el Culmen Depth
```

```python
# Flipper Length (mm)
_, mean_FL, std_FL, *_ = df2.FL.describe()
def get_z_score_FL(value, mean_FL, std_FL):
    return(value - mean_FL)/std_FL
df2["z_score_FL"] =df2.FL.apply(lambda x: get_z_score_FL(x,mean_FL,std_FL))
df2.head()
```

```python
df2[df2.z_score_FL>3]
```

```python
# No se encuentran valores cuyo Z-Score sea mayor a tres veces, por lo que no existen outliers para el Flipper Length
```

```python
# Modified Z-Score
# Body Mass
def get_mad(s):
    median = np.median(s)
    diff = abs(s-median)
    MAD = np.median(diff)
    return MAD
```

```python
MAD_BM = get_mad(df2.BM)
median=np.median(df2.BM)
MAD_BM, median
```

```python
def get_modified_z_score_BM (x, median, MAD_BM):
    return 0.6745*(x-median)/MAD_BM
```

```python
df2['mod_z_score_BM'] = df2.BM.apply(lambda x: get_modified_z_score_BM(x, median, MAD_BM))
df2.head()
```

```python
df2[df2.mod_z_score_BM>3.5]
```

```python
# No se tienen valores que sobrepasen 3.5 veces el Z Score mofificado del atributo Body Mass
```

```python
# Delta 15 N (o/oo)
MAD_N = get_mad(df2.N)
median=np.median(df2.N)
MAD_N, median
```

```python
def get_modified_z_score_N (x, median, MAD_N):
    return 0.6745*(x-median)/MAD_N
```

```python
df2['mod_z_score_N'] = df2.N.apply(lambda x: get_modified_z_score_N(x, median, MAD_N))
df2.head()
```

```python
df2[df2.mod_z_score_N>3.5]
```

```python
# No se tienen valores que sobrepasen 3.5 veces el Z Score mofificado del atributo Delta 15 N (o/oo)
```

```python
# Delta 13 C (o/oo)
MAD_C = get_mad(df2.C)
median=np.median(df2.C)
MAD_C, median
```

## <a id='5.2'>5.2 Comparación de distribuciones para atributos de comportamiento similar entre especíes.</a>

Dado que se establecen comportamientos muy simiares para algunos atributos entre dos especíes, se desea validar si son estadísticamente similares o no.

Para esto se utiliza la prueba Kolmogorov- Smirnov Goodness of fit, que permite comparar dos distribuciones para establecer si tienen una misma distribución:

Estadístico Kolmogorov - Smirnov para

    1. Culmen Length entre Chinstrap y Gentoo.
    2. Culmen Depth entre Adelie y Chinstrap.
    3. Flipper Length entre Adelie y Chinstrap.
